```python
import tensorflow as tf
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
Datadirectory="C:\\Users\\Moham\\Downloads\\archive\\train" #training dataset
```

```python
Classes=["0","1","2","3","4","5","6"] #list of classes
```

```python
for category in Classes:
    path=os.path.join(Datadirectory, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        plt.show
        break
    break
```

```python
img_size=224
new_array=cv2.resize(img_array, (img_size,img_size))
plt.imshow(cv2.cvtColor(new_array, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Read all the images and convert them into array
```

```python
training_data=[]
def create_training_data():
    for category in Classes:
        path=os.path.join(Datadirectory, category)
        class_num= Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array= cv2.resize(img_array, (img_size,img_size))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass
```

```python
create_training_data()
```

```python
print (len(training_data)) #number of images in training dataset
```

```python
import random
random.shuffle(training_data)
```

```python
x=[] #data
y=[] #label or class
for features, label in training_data:
    x.append(features)
    y.append(label)
x = np.array(x,dtype='uint8').reshape(-1, img_size, img_size, 3)
```

```python
y = np.array(y,dtype='uint8')
```

```python
#normalizing data
x= x/255.0
```

```python
######### Transfer Learning
#imports
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model=tf.keras.applications.MobileNetV2() ##pretrained model
```

```python
model.summary()
```

```python
# Transfer Learning, Fine tuning
```

```python
base_input= model.layers[0].input
```

```python
base_output= model.layers[-2].output
```

```python
base_output
```

```python
#adding final output layer
final_output= layers.Dense(128)(base_output)
final_output= layers.Activation('relu')(final_output)
final_output= layers.Dense(64)(final_output)
final_output= layers.Activation('relu')(final_output)
final_output= layers.Dense(7,activation='softmax')(final_output)
```

```python
final_output
```

```python
new_model= keras.Model(inputs= base_input, outputs= final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics= ["accuracy"])
```

```python
new_model.fit(x, y, epochs=20)
```

```python
new_model.save("C:\\Users\\Moham\\OneDrive\\Desktop\\AI\\emotion_classifier.h5")
```

```python
#testing the model
testing_data=[]
def create_testing_data():
    for category in Classes:
        path=os.path.join("C:\\Users\\Moham\\Downloads\\archive\\test", category)
        class_num= Classes.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array= cv2.resize(img_array, (img_size,img_size))
                testing_data.append([new_array, class_num])
            except Exception as e:
                pass
```

```python
create_testing_data()
```

```python
print (len(testing_data))
```

```python
random.shuffle(testing_data)
```

```python
x_test=[] #data
y_test=[] #label or class
for features, label in testing_data:
    x_test.append(features)
    y_test.append(label)
x_test = np.array(x_test,dtype='uint8').reshape(-1, img_size, img_size, 3)
```

```python
y_test = np.array(y_test,dtype='uint8')
```

```python
#normalizing data
x_test= x_test/255.0
```

```python
new_model=tf.keras.models.load_model("C:\\Users\\Moham\\OneDrive\\Desktop\\AI\\emotion_classifier.h5")
```

```python
new_model.evaluate(x_test,y_test)
```
